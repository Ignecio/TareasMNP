## Soluci√≥n

De igual manera que calculamos $ \pi $ en clases con la funcion arcotangente, lo hacemos aca con la funcion $ \sin^{-1} $. Por series de taylor sabemos que 

$$ \sin^{-1}(1/2) = \frac{\pi}{3} $$

Desarrollando la serie de Taylor de la arcoseno obtenemos:
$$ \pi = 6 \times ( \frac{1}{2}+ \frac{1}{4 \times 1 \times 5} \times \frac{1}{8}+ \frac{24}{16\times4\times5}\times \frac{1}{32}...) $$

```python
import matplotlib.pyplot as plt
import numpy as np

pi_arcsen = [2]
pi_arccos = [3]
pi_arctan = [4] # funcion vista en clase para calcular pi
x = 1
y = 1
for n in range(1, 100): 
    
    x=x*(2*n-1)
    y=y*(2*n)
    
    pi_arctan.append(pi_arctan[n-1] + (-1)**n*4/(2*n+1))
    pi_arcsen.append(pi_arcsen[n-1] + (2/(2*n+1))*x/y)
    pi_arccos.append(pi_arccos[n-1] + (-1)**(n+1)*4/((2*n)*(2*n+1)*(2*n+2)))
    
# ploteamos con loglog para ver la grafica con valores logaritmicos

plt.loglog(np.arange(100), pi_arcsen, label="arcsen", color="cyan")
plt.loglog(np.arange(100), pi_arccos, label="arccos", color="green")
plt.loglog(np.arange(100), pi_arctan, label="arctan", color="orange")
plt.loglog(np.arange(100), np.full(100, np.pi), label="pi verdadero", color= "purple")
plt.legend()

plt.ylim(np.pi - 0.1 * np.pi, np.pi + 0.1 * np.pi) # limitamos el eje y para poder apreciar de mejor manera la gr√°fica.
```

Nos acercamos aun mas a la gr√°fica ya que no logramos apreciar bien la diferencia entre el pi verdadero y el calculado por arccos.

```python
plt.loglog(np.arange(100), pi_arcsen, label="arcsen", color="cyan")
plt.loglog(np.arange(100), pi_arccos, label="arccos", color="green")
plt.loglog(np.arange(100), pi_arctan, label="arctan", color="orange")
plt.loglog(np.arange(100), np.full(100, np.pi), label="pi verdadero", color= "purple")
plt.legend()
plt.ylim(np.pi - 0.001 * np.pi, np.pi + 0.001 * np.pi) # nos acercamos mas.
```

Se puede notar que la aproximacion de $ \sin^{-1} $ es mala ya que nisiquiera se acerca al valor verdadero, en cambio la aproximaci√≥n de $ \cos^{-1} $ hacia el numero $\pi$ es mas notoria a medida de que n se vuelve mayor, mientras que en la funcion del arctangente no es muy notoria la eficiencia a medida que crece n. Claramente de los elementos mayores de cada sumatoria, el mas cercano es el de la funci√≥n del coseno.
